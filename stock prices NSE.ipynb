```python
#1. Moving average for weeks

def moving_average(stock,week=[4,16,28,40,52]):
    m_a = stock['Close'].rolling(week).mean()
    return (m_a)
   

infy_nse_m_a = moving_average(infy_nse,4)
tcs_nse_m_a = moving_average(tcs_nse,4)
nifty_it_m_a = moving_average(nifty_it,4)
```

```python
#2.rolling window for 10 and 75
def rolling_window(stock,window):
    roll_win = stock['Close'].rolling(window).mean()
    return (roll_win)

```

```python
#Volume shocks
def volume_shocks(stock):
    stock["vol_next_day"] = stock.Volume.shift(1)  
    
    stock["Volume_difference"] = 100*(stock["vol_next_day"] - stock["Volume"])/stock["Volume"]
    stock["volume_shock"] = [1 if abs(i)>10 else 0 for i in stock["Volume_difference"]]
    
    return stock
```

```python
nifty.head(2)
```

```python
#direction of shock
def direction_shock(stock1):
    stock1["vol_shock_direction"] = [1 if i >0 else 0 for i in stock1["volume_shock"]]
    return (stock1)
```

# price shock


```python
#for infy and tcs
def Price_Shock(stock2):
    stock2['price_difference'] = 100*(stock2['Close']-stock2['Prev Close']/stock2['Prev Close'])
    stock2['price_shock'] = [1 if abs(i)>2 else 0 for i in stock2['price_difference']]
    return stock2
```

```python
Price_Shock(infy_ns)
```

```python
Price_Shock(tcs_ns)
```

```python
#for nifty
nifty['close_next_day'] = nifty.Close.shift(1)
nifty.head(2)
```

```python
def Price_shock_nifty(stock3):
    stock3['price_difference_nifty']= 100*(stock3['close_next_day']-stock3['Close']/stock3['Close'])
    stock3['price_shock'] = [1 if abs(i)>2 else 0 for i in stock3['price_difference_nifty']]
    return stock3
```
